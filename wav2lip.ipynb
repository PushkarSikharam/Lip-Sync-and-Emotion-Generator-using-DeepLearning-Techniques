```python
%%writefile app.py
import streamlit as st
import uuid
import os
from pathlib import Path

import argparse
import subprocess
import uuid

def run_emotion_generation(face, audio, outfolder):
    try:

        # Construct the command for the subprocess
        command = [
            "python", "/content/emotalkingface/generate.py",  # Replace 'your_script_name.py' with the actual script name
            "--model", '/content/emotalkingface/model/',
            "--pred-path", '/content/emotalkingface/data/dlib_data/shape_predictor_68_face_landmarks.dat',
            "--temp-img", '/content/emotalkingface/data_prep/template_face.jpg',
            "--img-file", face,
            "--speech-file", audio,
            "--out-fold", outfolder
        ]

        # Run the command
        subprocess.run(command, check=True)

        d = Path(outfolder).iterdir()
        d = list(filter(lambda x:'mp4' in str(x),d))
        d = {i.stem.split('_')[0]:str(i) for i in d}
        print(d)
        # Return the output video file path
        return d
    except Exception as e:
        return f"Error: {str(e)}"

# Function to call the script with provided arguments
def run_lip_sync(face, audio, outfile):
    try:

        # Construct the command for the subprocess
        command = [
            "python", "/content/Wav2Lip/inference.py",  # Replace 'your_script_name.py' with the actual script name
            "--checkpoint_path", '/content/wav2lip_gan.pth',
            "--face", face,
            "--audio", audio,
            "--outfile", outfile
        ]

        # Run the command
        subprocess.run(command, check=True)

        # Return the output video file path
        return "/content/temp/result.avi"
    except Exception as e:
        return f"Error: {str(e)}"

# Streamlit app
st.title("Video and Audio Processing App")

user_choice = st.selectbox(label = 'select', options = ['lipsync','emotiongen'], index=0)
if user_choice == 'lipsync':
    # File uploaders for video and audio
    video_file = st.file_uploader("Upload Video", type=["mp4", "avi"])
    audio_file = st.file_uploader("Upload Audio", type=["mp3", "wav"])

    # Process button
    if st.button("Process"):
        if not video_file:
            st.error("Please provide a video.")
        elif not audio_file:
            st.error("Please provide an audio file.")
        else:
            # Save uploaded files temporarily
            video_path = f"temp_{uuid.uuid4().hex}_{video_file.name}"
            audio_path = f"temp_{uuid.uuid4().hex}_{audio_file.name}"

            with open(video_path, "wb") as f:
                f.write(video_file.getbuffer())

            with open(audio_path, "wb") as f:
                f.write(audio_file.getbuffer())

            with st.spinner(text="In progress..."):
                try:
                    # Generate output file name
                    outputfile = uuid.uuid4().hex + ".mp4"

                    # Run lip sync function
                    new_video_path = run_lip_sync(video_path, audio_path, outputfile)
                    resolved_path = str(Path(new_video_path).resolve())
                    output_path = str(Path(new_video_path).parent / Path(new_video_path).name)

                    print(f"Generated video path: {resolved_path}")

                    # Check if the generated file exists before processing
                    if not os.path.exists(resolved_path):
                        raise FileNotFoundError(f"Lip sync output file not found: {resolved_path}")

                    # Prepare and run ffmpeg command
                    command = f"ffmpeg -y -i {outputfile} -vcodec libx264 output.mp4"
                    try:
                        subprocess.run(command, check=True, shell=True)
                    except subprocess.CalledProcessError as e:
                        st.error(f"FFmpeg processing failed: {e}")
                        raise e

                    # Check if the ffmpeg output file exists
                    if os.path.exists(outputfile):
                        st.success("Processing complete!")
                        st.video('output.mp4')
                    else:
                        st.error("An error occurred during FFmpeg processing.")

                except Exception as e:
                    st.error(f"An unexpected error occurred: {e}")

            # Clean up temporary files
            os.remove(video_path)
            os.remove(audio_path)
else:
    # File uploaders for video and audio
    image_file = st.file_uploader("Upload image", type=["jpg", "jpeg", "png"])
    audio_file = st.file_uploader("Upload Audio", type=["mp3", "wav"])

    # Process button
    if st.button("Process"):
        if not image_file:
            st.error("Please provide a Image.")
        elif not audio_file:
            st.error("Please provide an audio file.")
        else:
            # Save uploaded files temporarily
            image_path = f"temp_{uuid.uuid4().hex}_{image_file.name}"
            audio_path = f"temp_{uuid.uuid4().hex}_{audio_file.name}"

            with open(image_path, "wb") as f:
                f.write(image_file.getbuffer())

            with open(audio_path, "wb") as f:
                f.write(audio_file.getbuffer())

            with st.spinner(text="In progress..."):
                outputfolder = uuid.uuid4().hex
                res = run_emotion_generation(image_path, audio_path, outputfolder)
                print('*'*30)
                print('res',res)
                for key, path in res.items():
                    resolved_path = str(Path(path).resolve())
                    output_path = str(Path(path).parent / Path(path).name)

                    print(f"Processing file: {resolved_path}")
                    if not os.path.exists(resolved_path):
                        raise FileNotFoundError(f"File not found: {resolved_path}")

                    command = f"ffmpeg -y -i {resolved_path} -vcodec libx264 {output_path}"
                    try:
                        subprocess.run(command, check=True, shell=True)
                        st.write(key)
                        st.video(output_path)
                    except subprocess.CalledProcessError as e:
                        st.error(f"FFmpeg processing failed: {e}")

            # Clean up temporary files
            os.remove(image_path)
            os.remove(audio_path)
```
